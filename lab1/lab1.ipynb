```python
#Task: swap channels (rgb->rbg and etc.)
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import matplotlib.colors as pltcolor
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(20,15))
rgb = pltcolor.ListedColormap([(1,0,0),(0,1,0),(0,0,1)],name = 'list_rgb')
rbg = pltcolor.ListedColormap([(1,0,0),(0,0,1),(0,1,0)],name = 'list_rbg')
grb = pltcolor.ListedColormap([(0,1,0),(1,0,0),(0,0,1)],name = 'list_grb')
gbr = pltcolor.ListedColormap([(0,1,0),(0,0,1),(1,0,0)],name = 'list_gbr')
bgr = pltcolor.ListedColormap([(0,0,1),(0,1,0),(1,0,0)],name = 'list_bgr')
brg = pltcolor.ListedColormap([(0,0,1),(1,0,0),(0,1,0)],name = 'list_brg')
plt.subplot(321)
plt.imshow(im,cmap=rgb)
plt.subplot(322)
plt.imshow(im,cmap=rbg)
plt.subplot(323)
plt.imshow(im,cmap=grb)
plt.subplot(324)
plt.imshow(im,cmap=gbr)
plt.subplot(325)
plt.imshow(im,cmap=bgr)
plt.subplot(326)
plt.imshow(im,cmap=brg)

plt.show()
```

```python
#Task: Add cat's head somewhere else
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(211)
cat_head = im[200:600,800:1250]
merge = im
for i in range(0, 400):
    for j in range(0, 450):
        for k in range(0,3):
            merge[i,j,k] = cat_head[i,j,k]
plt.imshow(merge)
plt.show()
```
