Construct a histogram of the brightness of the image. Apply the operation "linear stretching of brightness" and "the gray world" to the image. Operation "the gray world" needs to be realized independently. The mean can be called three times.


```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
import skimage 
from matplotlib import image as pltimg
import scipy.ndimage as spimg
img = pltimg.imread('.\\ex1.jpg');
aver_r = np.mean(img[:,:,0])
aver_g = np.mean(img[:,:,1])
aver_b = np.mean(img[:,:,2])
aver_total = (aver_r+aver_g+aver_b)/3
img_gray = img
img_gray[:,:,0] = img[:,:,0]*(aver_total/aver_r)
img_gray[:,:,1] = img[:,:,1]*(aver_total/aver_g)
img_gray[:,:,2] = img[:,:,2]*(aver_total/aver_b)
img_bright = (img_gray-np.min(img_gray))*((255-0)/(np.max(img_gray)-np.min(img_gray)))
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(img)
plt.title('original image')
plt.subplot(222)
hist,bins = np.histogram(img,256,[0,256])
plt.hist(img.ravel(),256,[0,256])
plt.title('histogram')
plt.subplot(223)
plt.imshow(img_bright, cmap='gray')
plt.title('brighter image')
plt.subplot(224)
plt.imshow(img_gray)
plt.title('gray world')
plt.show()
```
